```python
def insert_tree_node(tree, node, metadata):
    """
    Hàm đệ quy để chèn một node từ cấu trúc 'tree' vào cây chính.
    Node cha cũng sẽ chứa metadata từ cây con trong metadata.
    """
    current = tree
    name = node['name']
    
    # Nếu node đã tồn tại, tiếp tục đi sâu vào cây con
    if name not in current:
        current[name] = {
            'metadata': {
                'name': node['name'],
                'display_name': node.get('display_name'),
                'description': node.get('description'),
                'parrent': node.get('parrent')
            },
            'children': {},
            'leaf_count': 0  # Sẽ dùng để đếm các node lá
        }
    
    # Nếu node có con, tiếp tục đệ quy với cây con
    if 'children' in node and node['children']:
        for child in node['children']:
            insert_tree_node(current[name]['children'], child, metadata)
    else:
        # Nếu node không có con, đó là node cuối cùng -> thêm metadata
        if 'list_data' not in current[name]:
            current[name]['list_data'] = []
        current[name]['list_data'].append(metadata)

def count_all_leaf_nodes(tree):
    """
    Hàm đệ quy để đếm tổng số node lá trong cây (những node cuối cùng không có con).
    """
    total_leaves = 0
    
    for node, value in tree.items():
        # Nếu node có con, đệ quy vào cây con để đếm tổng số lá của cây đó
        if value['children']:
            value['leaf_count'] = count_all_leaf_nodes(value['children'])
        else:
            # Nếu là node lá, gán leaf_count là 1
            value['leaf_count'] = 1
        
        # Cộng dồn số node lá
        total_leaves += value['leaf_count']
    
    return total_leaves

# Dữ liệu đầu vào
data = sources
# Cấu trúc cây dữ liệu
tree = {}

# Chèn dữ liệu vào cây
for entry in data:
    insert_tree_node(tree, entry['metadata']['tree'], entry['metadata'])

# Đếm tổng số node lá
count_all_leaf_nodes(tree)

import json
print(json.dumps(tree, indent=4))
```

```python
def insert_tree_node(tree, node, metadata):
    """
    Recursive function to insert a node from 'tree' structure into the main tree.
    If it's a leaf node, metadata is added to list_data.
    """
    current = tree
    name = node['name']

    # Create a new node if it does not exist
    if name not in current:
        current[name] = {
            'metadata': {
                'name': node['name'],
                'display_name': node.get('display_name'),
                'description': node.get('description'),
                'parrent': node.get('parrent')
            },
            'children': {},
            'leaf_count': 0,
            'list_data': []  # Ensure list_data exists for adding leaf metadata directly
        }

    # If children are present, recursively add child nodes
    if 'children' in node and node['children']:
        for child in node['children']:
            insert_tree_node(current[name]['children'], child, metadata)
    else:
        # If it is a leaf node, append metadata to list_data at the current level
        current[name]['list_data'].append(metadata)


def count_all_leaf_nodes(tree):
    """
    Recursive function to count the total number of leaf nodes and assign leaf_count to parent nodes.
    """
    total_leaves = 0
    for node, value in tree.items():
        # If children exist, recurse into the subtree to count leaf nodes
        if value['children']:
            value['leaf_count'] = count_all_leaf_nodes(value['children'])
        else:
            # If it's a leaf node, set leaf_count to 1
            value['leaf_count'] = 1

        # Accumulate the count of leaf nodes
        total_leaves += value['leaf_count']

    return total_leaves


# Input data
data = sources

# Initialize the main data tree
tree = {}

# Insert data into the tree
for entry in data:
    insert_tree_node(tree, entry['metadata']['tree'], entry['metadata'])

# Count total leaf nodes
count_all_leaf_nodes(tree)

# Print final structure
import json
print(json.dumps(tree, indent=4, ensure_ascii=False))
```
